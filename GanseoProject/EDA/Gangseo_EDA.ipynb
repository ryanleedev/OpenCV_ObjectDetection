```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import plotly.express as px
df_raw  = pd.read_csv('gangseo_accident_EDA.txt' ,  sep="\t" , encoding = 'utf-8')
```

```python
df_raw
```

```python
df = df_raw.iloc[:,2:]
df
```

```python
#차대차 데이터
df_ctoc = df.loc[df.사고유형 == '차대차']
df_ctoc
```

```python
#차대사람 데이터
df_ctoh = df.loc[df.사고유형 == '차대사람']
df_ctoh
```

```python
#차대차 법정동별 사고건수 카운트
df_ctocC = df_ctoc.법정동.value_counts()
df_ctocC
```

```python
#차대사람 법정동별 사고건수 카운트
df_ctohC = df_ctoh.법정동.value_counts()
df_ctohC
```

```python
#동별 차대차/차대사람 사고건수
df1 = df[['법정동','사고유형']]
df1
fig = px.bar(df1, x="법정동" , color="사고유형", title="각 동별 차대차, 차대사람 사고건수")
fig.show()
```

```python
# 동별 통합
df['법정동'] = df.법정동.str.replace('1','')
df['법정동'] = df.법정동.str.replace('2','')
df['법정동'] = df.법정동.str.replace('3','')
df['법정동'] = df.법정동.str.replace('4','')
df['법정동'] = df.법정동.str.replace('5','')
df['법정동'] = df.법정동.str.replace('6','')
df['법정동'] = df.법정동.str.replace('7','')
df['법정동'] = df.법정동.str.replace('8','')
df['법정동'] = df.법정동.str.replace('본','')
df
```

```python
#동별(통합) 사고(통합) 건수
fig = px.bar(df, x="법정동" , color="사고유형", title="동별 차대차, 차대사람 사고건수")
fig.show()
```

```python
#동별(통합) 차대사람 사고 데이터
df_ctoh2 = df.loc[df.사고유형 == '차대사람']
df_ctoh2
```

```python
#동별(통합) 차대사람 사고 건수
df_ctoh2dong = df_ctoh2['법정동'].value_counts()
df_ctoh2d= df_ctoh2dong.reset_index(False) 
df_ctoh2d.columns = ['법정동', '사고건수']
df_ctoh2d
```

```python
#동별(통합) 인구밀도
afaf = pd.read_csv('popularity.csv', encoding='euc=kr')
afaf.columns = ['법정동','인구','면적','인구밀도']
afaf[['법정동','인구밀도']]
```

```python
#동별(통합) 차대사람 인구밀도 merge
df_cthd = pd.merge(left = df_ctoh2d , right = afaf, how = "outer", on = "법정동")
df_cthd.인구밀도 = df_cthd.인구밀도.str.replace(',','')
df_cthd.인구밀도 = pd.to_numeric(df_cthd.인구밀도)
df_cthd.info()
df_cthd
```

```python
import plotly.express as px

fig = px.bar(df_cthd, x='법정동', y='사고건수',
             hover_data=['인구', '인구밀도'], color='인구밀도',
             labels={'count':'차대사람 사고건수'}, height=800, color_discrete_sequence=px.colors.qualitative.D3, title='동별 차대사람 사고건수').update_xaxes(categoryorder="total descending")
fig.show()
```
